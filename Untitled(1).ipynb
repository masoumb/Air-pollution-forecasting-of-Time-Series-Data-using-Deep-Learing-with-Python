### Insights from heat-map Plot

2. "pdays&deposit" features: Pdays(number of days that passed by after the client was last contacted from a previous campaign) has a posite correlation with the ter deposit. This shows that i is better to give sutomers some time before starting with new campaing and contacts. 

3. Previous (number of contacts before the current campaign) also has a positive correlation with the number of clients subscribed to the term deposit.

4. Campaign which is the number of contacts in the current campaign has a negative correlation with the term deposit. Increased number of contacts with customers doesn't seem to have a positive result in the subscription to the term deposit.
